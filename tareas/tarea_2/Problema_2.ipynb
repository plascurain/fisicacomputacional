# Problema 2

Queremos enocontrar las raíces de la función: 

$$ f(x,y) = 2e^{-(x+1)^2} \left ( e^{-(y+2)^2} + e^{-(y-2)^2} \right ) + 6x(x^2+y^2)e^{-(x^2+y^2)} - 1 $$

Voy a definir a la función como una función aparentemente de 1 variable, pero en realidad nuestra variable "r" es un array de dos entradas "r=[[x],[y]]"; además, de esta manera es más agradable a la vista la definición de la función. 

Para que quede más claro, voy a definir los argumentos de las exponenciales de la siguiente manera: 

arg_1x = $-(x+1)^2$

arg_2y = $-(y+2)^2$

arg_3y = $-(y-2)^2$

arg_4xy = $-(x^2+y^2)$

```python
def p2(r): 
    arg_1x = (r[0] + 1.)**2
    arg_2y = (r[1] + 2.)**2
    arg_3y = (r[1] - 2.)**2
    arg_4xy = r[0]**2 + r[1]**2
    return 2.*exp(-arg_1x)*(exp(-arg_2y) + exp(-arg_3y)) + 6.*r[0]*arg_4xy*exp(-arg_4xy)-1
```

```python
fig = plt.figure(figsize=(20,10))

ax = fig.add_subplot(241)
bx = fig.add_subplot(242) 
cx = fig.add_subplot(243)
dx = fig.add_subplot(244)

ax.set_title("tol 0.1")
bx.set_title("tol 0.08")
cx.set_title("tol 0.05")
dx.set_title("tol 0.01")

#0.1
raices_x1, raices_y1 =  todas_raices_2d(p2,[-4,4,-4,4],0.1)
ax.plot(raices_x1,raices_y1,'.',color = 'y')

#0.08
raices_x2, raices_y2 =  todas_raices_2d(p2,[-4,4,-4,4],0.08)
bx.plot(raices_x2,raices_y2,'.',color = 'r')

#0.05
raices_x3, raices_y3 =  todas_raices_2d(p2,[-4,4,-4,4],0.05)
cx.plot(raices_x3,raices_y3,'.',color = 'g')

#0.01
raices_x4, raices_y4 =  todas_raices_2d(p2,[-4,4,-4,4],0.01)
dx.plot(raices_x4,raices_y4,'.',color = 'b')

show()
#es una carita feliz
```
