# El problema de Keppler

#### Con $epsilon = 1$

El espacio fase es aquel donde hay constantes de movimiento

```python
from pylab import*
from scipy.integrate import*
%matplotlib inline
```

```python
#x_p[0]=x, x_p[1]=y, x_p[2]=px, x_p[3]=py
def ec_mov(x_p,t,epsilon=1.):
    r32=(x_p[0]**2+x_p[1]**2)**(1.5)
    dx = x_p[2]
    dy = epsilon*x_p[3]
    ddx = -x_p[0]/r32
    ddy = -x_p[1]/r32
    return array([dx,dy,ddx,ddy])

def energia(x_p,epsilon=1.):
    r = (x_p[0]**2+x_p[1]**2)**(0.5)
    E = 0.5*(x_p[2]**2+epsilon*x_p[3]**2) - 1./r
    return E

def cond_ini(alfa,var=0.001):
    angulo = alfa+var*(2.*rand()-1.)
    x,y = cos(angulo),sin(angulo)
    return array([x,y,1,0.])

def p_rho(x_p):
    r = (x_p[:,0]**2+x_p[:,1]**2)**(0.5)
    x,y,px,py=x_p[:,0],x_p[:,1],x_p[:,2],x_p[:,3]
    p_r = (x*px + y*py)/r
    return p_r,r
```

```python
t = linspace(0,10,300)
for i in range(10):
    p_ini = [cos(i),sin(i),.5,.5]
    sol = odeint(ec_mov,p_ini,t,args=(1.001,))
    plot(sol[:,0],sol[:,1])
    print "E= ", energia(p_ini),"E_f= ", energia(sol[-1])
show()
```

```python
def root_trayec1(fun,p_i,ti,error=1e-5):
    while abs(p_i[0])> error:
        bla = ti -(p_i[0]/p_i[2])
        paso = odeint(fun,p_i,[ti,bla])
        ti,p_i=bla,paso[-1]
    return ti,p_i
```
