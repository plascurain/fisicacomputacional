### Definición de  los métodos de integración: 

### a) Método de Euler

Primero vamos a definir nuestra función y condiciones iniciales: 
$$ \frac{dy}{dt} = f(t,y) \\\\ y(t_0)=y_0$$

Partimos de que la tangente a la solución que conocemos está dado por 
$$ y=y_0+f(t_0,y_0)(t-t_0)$$ 

Para el caso de $y_1$: 
$$ y_1=y_0+f(t_0,y_0)(t_1-t_0)$$ 

Y si pensamos que el punto consecuente $y_1$ está lo suficientemente cerca como para aproximar la pendiente entonces podemos calcular de la misma forma el punto $y_2$. Si repetimos ese paso llegaremos a la fórmula más general: 

$$ y_{n+1}=y_n+f(t_n,y_n)h$$ 
Donde $h=t_{n+1}-t_n$

```python
def euler(ec, y_ini, x, h = 0.001):
    
    tiempo = arange(x[0],x[-1] + h, h)
    sol = zeros((len(tiempo),len(y_ini)))
    sol[0,:] = y_ini
    
    for i in xrange(len(tiempo)-1):
            sol[i+1] = sol[i] + ec(sol[i],tiempo[i])*h
            
    return tiempo, sol  
```

### b) Método de Taylor de 2do orden

Si tenemos que $y'=f(x,y)$ y que $y(x_0)=y_0$ y queremos analizar nuestro método en un intervalo $x_0+h$ con $h$ el paso de integración...

$$ y(x_0+h) = y_0 + hy'(x_0) +\frac{h^2}{2} y''(x_0) + O(h^3)$$ 
donde $O(x_0)$ serán los términos de orden superior ya que truncamos la serie en el segundo orden.



```python

def taylor(ec,d_ec,y_ini,x,h=0.1):
    if abs(x[1]-x[0])<=10.*h:
        h = abs(x[1]-x[0])/100
    tiempos = arange(x[0],x[1]+h,h)
    sol = zeros((len(tiempos),len(y_ini)))
    sol[0,:] = y_ini
    for i in xrange(len(tiempos)-1):
        f=ec(sol[i,:],tiempos[i])
        df,jac=d_ec(sol[i,:],tiempos[i])
        sol[i+1,:]=sol[i,:]+h*f+(0.5*h*h*(jac[0] + dot(jac[1],f)))
    return tiempos,sol
```

### c) Método de Runge-Kutta de 2do orden


```python
def rk2(ec,y_ini,x,h=0.1):
    tiempos = arange(x[0],x[1]+h,h)
    sol = zeros((len(tiempos),len(y_ini)))
    sol[0,:] = y_ini
    for i in xrange(len(tiempos)-1): 
        k = h*ec(sol[i,:],tiempos[i])
        f = ec(sol[i,:]+0.5*k,tiempos[i]+0.5*h) #la ec.dif evaluada en la primera entrada y tiempos
        sol[i+1,:]=sol[i,:]+h*f
    return tiempos,sol
```

### d) Método de Runge-Kutta de 4to orden

$y_{n+1} = y_n +\frac{h}{6}(k_1+2k_2+2k_3+k_4)$

$k_1=f(x_n,y_n)$

$k_2= f(x_n+\frac{h}{2},y_n+\frac{h}{2}k_1)$

$k_3 = f(x_n +\frac{h}{2},y_n + \frac{h}{2}k_2)$

$k_4 = f(x_n +h,y_n + hk_3)$

```python
def rk4(ec,y_ini,x,h=0.1):
    tiempos = arange(x[0],x[1]+h,h)
    sol = zeros((len(tiempos),len(y_ini)))
    sol[0,:] = y_ini
    for i in xrange(len(tiempos)-1): 
        k1 = ec(sol[i,:],tiempos[i])
        k2 = ec(sol[i,:]+0.5*h*k1,tiempos[i]+0.5*h) 
        k3 = ec(sol[i,:]+0.5*h*k2,tiempos[i]+0.5*h) 
        k4 = ec(sol[i,:]+h*k3,tiempos[i]+h) 
        sol[i+1,:]=sol[i,:]+(h/6.)*(k1+2*k2+2*k3+k4) 
    return tiempos,sol
```

## Problema 1:

Vamos primero a hacer una prueba de funcionamiento con los cuatro integradores usando la ecuación diferencial: 

$$ \ddot{x} + 0.5 \dot{x} + \sin{x} = e^{-t} \sin{3t}$$

Y vamos a separarlo en un sistema de ecuaciones: 
$$\dot{x} = y$$

$$ \dot{y} = e^{-t} \sin{3t} - 0.5 \dot{x} - \sin{x}$$

con las condiciones iniciales: $\dot{x}(0)=1$ & $x(0)=0$

```python
def f_prueba(y,t):
    return array([y[1], exp(-t)*sin(3*t) -0.5*y[1] - sin(y[0])])

def d_prueba(x,t):
    par = array([0,exp(-t)*(3*cos(3*t)-sin(3*t))])
    jac = array([ [0, 1], [-cos(x[0]), -0.5] ])
    return par,jac
```

## Problema 2 

Queremos encontrar el paso más pequeño para resolver la siguiente ecuación diferencial: 

$$ \dddot{y} + \ddot{y}^2 -3\dot{y}^3 + \cos^2(y) = e^{-t} \sin(3t) $$ 

Con las condiciones iniciales: 

$\ddot{y}(1)=1$ 

$\dot{y}(1)=2$

$y(1)=1$

Primero vamos a hacer el siguiente cambio de variable: 

$\dot{y}=u$ & $\dot{u}=s$

Entonces la ecuación diferencial queda como: 

$$ \dot{s} + s^2 -3u^3 + \cos^2(y) = e^{-t} \sin(3t) $$

Así tenemos el sistema de ecuaciones: 

$\dot{y}=u$

$\dot{u}=s$

$\dot{s} = -s^2 +3u^3 - \cos^2(y) + e^{-t} \sin(3t)$

```python
def fp2(r,t): 
    y = r[0]
    u = r[1]
    s = r[2]
    y_p = r[1]
    u_p = r[2]
    s_p = -r[2]**2 + 3*r[1]**3 - (cos(r[0]))**2 + exp(-t)*sin(3*t)
    
    return array([y_p,u_p,s_p])
```

```python
def d_fp2(r,t):
    return array([0,0,(3*cos(3*t)-sin(3*t))*exp(-t)]), array([[0,1,0], [0,0,1], [sin(2*r[0]),9*r[1]*r[1],-2*r[2]]])
```

## Problema 3: 

Tenemos la ecuación diferencial: 

$$ \ddot{x} + \frac{1}{10}\dot{x} + 4sin(x) = \frac{1}{2} sin(\omega t) $$

Con condiciones iniciales : $\dot{x}(0)=1$ ; $x(0)=0$ y para $\omega = [1,1.1,1.2,...,2.9,3]$

Esta ecuación la vamos a solucionar haciendo el cambio de variable: 

$$ \dot{x} = u $$

$$ \dot{u} = \frac{1}{2} sin(\omega t) - \frac{1}{10}u - 4sin(x)$$

Tenemos también que : 

$$ \frac{\partial f}{\partial t} = (0,\frac{1}{2}\omega cos(\omega t))$$

Y por último, el Jacobiano es: 

$$ 
  J(x,u)=  \begin{bmatrix}
    0 & 1 \\
    -4cos(x) & -\frac{1}{10} 
  \end{bmatrix}
 $$

Primero vamos a hacer las modificaciones para los integradores de la pregunta 1 tal que puedan aceptar parámetros extras 


### Integradores modificados: 

```python
def euler_m(ec, p_ini, x, args, h=0.01):
    tiempos = arange(x[0],x[1]+h,h)
    y = zeros((len(tiempos),len(p_ini)))
    y[0,:] = p_ini
    for i in range(len(y)-1):
        y[i+1] = y[i] + h*ec(y[i], tiempos[i], args)
    return tiempos, y

def taylor_m(ec,d_ec,p_ini,x,args,h=0.1):
    tiempos = arange(x[0],x[1]+h,h)
    sol = zeros((len(tiempos),len(p_ini)))
    sol[0,:] = p_ini
    for i in xrange(len(tiempos)-1):
        f = ec(sol[i,:],tiempos[i], args)
        df,jac = d_ec(sol[i,:],tiempos[i], args)
        sol[i+1,:]=sol[i,:]+h*f+(0.5*h*h*(df+dot(jac,f)))
    return tiempos,sol

def rk2_m(ec, p_ini, x, args, h=0.1):
    tiempos = arange(x[0],x[1]+h,h)
    sol = zeros((len(tiempos),len(p_ini)))
    sol[0,:] = p_ini
    for i in xrange(len(tiempos)-1): 
        k = h*ec(sol[i,:],tiempos[i], args)
        f = ec(sol[i,:]+0.5*k,tiempos[i]+0.5*h, args)
        sol[i+1,:] = sol[i,:]+h*f
    return tiempos,sol

def rk4_m(ec, p_ini, x, args, h=0.1):
    tiempos = arange(x[0],x[1]+h,h)
    sol = zeros((len(tiempos),len(p_ini)))
    sol[0,:] = p_ini
    for i in xrange(len(tiempos)-1): 
        k1 = ec(sol[i,:],tiempos[i], args)
        k2 = ec(sol[i,:]+0.5*h*k1,tiempos[i]+0.5*h, args)
        k3 = ec(sol[i,:]+0.5*h*k2,tiempos[i]+0.5*h, args)
        k4 = ec(sol[i,:]+h*k3,tiempos[i]+h, args)
        sol[i+1,:] = sol[i,:] + h*(k1+2*k2+2*k3+k4)/6
    return tiempos,sol
```

Vamos a definir ahora la ecuación diferencial y su jacobiano (los que ya calculamos)

```python
def fp3(x, t, args):
    X_prime = x[1]
    Y_prime = -0.1*x[1] - 4.0*sin(x[0]) + 0.5*sin(args[0]*t)
    return array([X_prime, Y_prime])

def d_fp3(x, t, args):
    return array([0, 0.5*args[0]*cos(args[0]*t)]), array([[0,1],[-4*cos(x[0]),-0.1]])
```

Lo que vamos a hacer para cada uno de los integradores es: 
1.- definir un intervalo de tiempo en el que vamos a integrar
2.- dar un número máximo de las soluciones que queremos ver
3.- hacer dos arreglos para meter ahí todas las soluciones
4.- calcular y graficar

### Método de Euler

```python
Tmax = 50

n=5 #tomamos 5 soluciones (contando desde 0)
w = [1.0 + (3.0-1.0)*i/(n-1) for i in range(n)] #los valores que va a ir tomando omega


t_soluciones = []
soluciones = []
for i in range(n):
    t_e3, sol_e3 = euler_m(fp3, array([0.0, 1.0]), [0, Tmax], [w[i]], 0.01)
    t_soluciones.append(t_e3)
    soluciones.append(sol_e3)
    plot(t_soluciones[i],soluciones[i][:,0],label =i ) 
plt.title('Euler')    
legend()
show()
```

### Método de Taylor

```python
Tmax = 50

n=5
w = [1.0 + (3.0-1.0)*i/(n-1) for i in range(n)]

t_soluciones = []
soluciones = []

for i in range(n):
     
    t_t3, sol_t3 = taylor_m(fp3,d_fp3,array([0.0, 1.0]), [0, Tmax], [w[i]], 0.01)
    t_soluciones.append(t_t3)
    soluciones.append(sol_t3)
    plot(t_soluciones[i],soluciones[i][:,0],label =i )
plt.title('Taylor')    
legend()
show()
```

### Método de Runge-Kutta 2

```python
Tmax = 50

n=5
w = [1.0 + (3.0-1.0)*i/(n-1) for i in range(n)]

t_soluciones = []
soluciones = []

for i in range(n):
     
    t_rk23, sol_rk23 = rk2_m(fp3,array([0.0, 1.0]), [0, Tmax], [w[i]], 0.01)
    t_soluciones.append(t_rk23)
    soluciones.append(sol_rk23)
    plot(t_soluciones[i],soluciones[i][:,0],label =i )
plt.title('RK-2')    
legend()
show()
```

### Método de Runge-Kutta 4

```python
Tmax = 50

n=5
w = [1.0 + (3.0-1.0)*i/(n-1) for i in range(n)]

t_soluciones = []
soluciones = []

for i in range(n):
     
    t_rk43, sol_rk43 = rk4_m(fp3,array([0.0, 1.0]), [0, Tmax], [w[i]], 0.01)
    t_soluciones.append(t_rk43)
    soluciones.append(sol_rk43)
    plot(t_soluciones[i],soluciones[i][:,0],label =i )
plt.title('RK-4')    
legend()
show()
```

Veamos que para cada una de las n's, el valor de la función se comporta cualitativamente igual en cada uno de los métodos (esto lo podemos ver porque el color de las gráficas para cada valor de la n es el mismo). Además el comportamiento a lo largo de la gráfica es oscilatorio con los mismos rangos en la gráfica.

## Problema 4

Tenemos el siguiente sistema de ecuaciones diferenciales:

$$ \frac{dx}{dt} = 10(x-y) $$
$$ \frac{dy}{dt} = x(28-z) $$
$$ \frac{dz}{dt} = xy - \frac{8}{3}z $$

Y vamos a reaizar la integración con el método de Runge-Kutta de cuarto orden.

```python
def fp4(r,t): 
    x = r[0]
    y = r[1]
    z = r[2]
    x_p = 10*(y-x)
    y_p = x*(28-z) - y
    z_p = x*y - (8./3.)*z
    
    return array([x_p,y_p,z_p])
```
