```python
from pylab import *
%run "mis_integradores.ipynb"
%matplotlib inline
```

```python
def osc(y,t,w02=1.):
    return array([y[1],-w02*sin(y[0])+0.1*sin(t)]) 
```

```python
t,sol=rk4(osc,[0.5,0.5],[0,10])
plot(sol[:,0],sol[:,1])
zoom = 2
xlim([-zoom,zoom]),ylim([-zoom,zoom])
axes().set_aspect('equal'),grid(True)
show()
```

```python
for i in range(10): #el nmero entre parntesis es la cantidad de trayectorias
    y0,u0 = pi*(2*rand()-1),0 #,4*(2*rand()-1)
    t,sol=rk4(osc,[y0,u0],[0,100])
    plot(sol[:,0],sol[:,1])
zoom = pi
xlim([-zoom,zoom]),ylim([-zoom,zoom])
axes().set_aspect('equal'),grid(True)
show()
```
