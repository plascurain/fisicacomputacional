```python
from pylab import *
from scipy.integrate import *
%matplotlib inline
```

```python
def oscilador(y_v , t , omega2=1.):
    #y_v=[y,v]
    dy = y_v[1]
    dv = -omega2*y_v[0]
    
    return array([dy,dv])
```

```python
t = linspace(0,10)
inicial = array([0.,1.])
sol = odeint(oscilador , inicial , t)
plot(t,sol[:,0],label = "y")
plot(t,sol[:,1],label = "$v_y$")
legend(),grid()
show()
```

```python
def root_trayec(funcion , pto_inicial , ti , error = 1e-5):
    while abs(pto_inicial[0]) > error:
        tip1 = ti - (pto_inicial[0]/pto_inicial[1])   #ti del punto inicial
        paso = odeint(funcion , pto_inicial , [ti,tip1])
        ti , pto_inicial = tip1 , paso[-1]
    return ti , pto_inicial
```

```python
indice = 10
ti , pto_inicial = t[indice] , sol[indice]
t_raiz , punto = root_trayec(oscilador , pto_inicial , ti)

print t_raiz , punto
```
