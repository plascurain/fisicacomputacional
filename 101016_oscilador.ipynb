```python
from pylab import *
from scipy.integrate import *
%matplotlib inline
```

```python
#y_v[0] = y ; y_v[1] = v
def oscilador(y_v,t,omega2=1.,mu=0):
    dy = y_v[1]
    dv = -omega2*y_v[0] - mu*y_v[1]
    
    return array([dy,dv])
```

```python
t = linspace(0,50,500)
inicio = [0,1]
sol = odeint(oscilador , inicio , t , args=(1.,0.1))
xlim([-1,1])
ylim([-1,1])
plot(sol[:,0],sol[:,1])
show()
```

```python
from JSAnimation import IPython_display
from matplotlib import animation
```

```python
figura = figure()
ejes = axes(xlim=(-1,1) , ylim=(-1,1))

trayectoria, = ejes.plot([],[], lw = 2) #lw es el grosor de la linea
punto, = ejes.plot([],[], lw = 2)

def init(): #funcion de borron y cuenta nueva
    trayectoria.set_data([],[])
    punto.set_data([],[])
    
    return trayectoria , punto,

y , v=sol[:,0] , sol[:,1]

def el_movimiento(i):
    
    trayectoria.set_data(y[:i+1] , v[:i+1])
    punto.set_data(y[i] , v[i])
    
    return trayectoria,punto,

animation.FuncAnimation(figura , el_movimiento , init_func=init,frames = 100) #aqui son 20 porque si no, se tarda... 
                                                                             #pero para que se vea bonito hay que 
                                                                             #darle como 200 o asi, igual se tarda mas
```
